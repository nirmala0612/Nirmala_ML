We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

temp_max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months=np.arange(12)
plt.plot(months,temp_max,'go--')
plt.plot(months,temp_min,'r')
plt.xlabel('Months')
plt.ylabel('Max-Min Temperatures')
plt.show()

def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,temp_max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,temp_min, [-40, 20, 0])

days = np.linspace(0, 12, num=365)

plt.plot(months, temp_max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'green')
plt.plot(months, temp_min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'purple')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
plt.show()
```

Charts to plot:
1. Create a pie chart presenting the male/female proportion
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd

titanic=pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
count=titanic['sex'].value_counts()
colors=['purple','pink']
plt.pie(count.values.tolist(),labels=count.index.values.tolist(),colors=colors,autopct='%.1f%%')
plt.show()

titanic=titanic.dropna(subset=['sex'])
colors={'male':'orange','female':'green'}
plt.scatter(titanic['fare'],titanic['age'],c=titanic['sex'].map(colors))
plt.show()
```
